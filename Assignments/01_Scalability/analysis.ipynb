```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
def load_data(file):
    """Returns a numpy array"""
    data = []
    for line in open(file):
        data += [float(line[:-2])]
    return np.asarray(data)
```

## Weak Scaling

```python
weak5 = load_data("parallel/weak_scalability/weak_scal5.dat")
weak6 = load_data("parallel/weak_scalability/weak_scal6.dat")
weak7 = load_data("parallel/weak_scalability/weak_scal7.dat")
```

### Real time execution as a function of the number of cores
The number of core ranges from 1 to 20.

```python
# N = 10^5
plt.plot(weak5)
plt.scatter(np.arange(weak5.shape[0]), weak5)

# N = 10^6
plt.plot(weak6)
plt.scatter(np.arange(weak6.shape[0]), weak6)

# N = 10^7
plt.plot(weak7)
plt.scatter(np.arange(weak7.shape[0]), weak7)

plt.show()
```

```python
weak5_speedup = weak5[0] / weak5
weak6_speedup = weak6[0] / weak6
weak7_speedup = weak7[0] / weak7
```

```python
# N = 10^5
plt.plot(weak5_speedup)
plt.scatter(np.arange(weak5_speedup.shape[0]), weak5_speedup)

# N = 10^6
plt.plot(weak6_speedup)
plt.scatter(np.arange(weak6_speedup.shape[0]), weak6_speedup)

# N = 10^7
plt.plot(weak7_speedup)
plt.scatter(np.arange(weak7_speedup.shape[0]), weak7_speedup)
```

```python
weak5_efficiency = weak5_speedup / np.arange(1, 21)
weak6_efficiency = weak6_speedup / np.arange(1, 21)
weak7_efficiency = weak7_speedup / np.arange(1, 21)
```

```python
# N = 10^5
plt.plot(weak5_efficiency)
plt.scatter(np.arange(weak5_efficiency.shape[0]), weak5_efficiency)

# N = 10^6
plt.plot(weak6_efficiency)
plt.scatter(np.arange(weak6_efficiency.shape[0]), weak6_efficiency)

# N = 10^7
plt.plot(weak7_efficiency)
plt.scatter(np.arange(weak7_efficiency.shape[0]), weak7_efficiency)
```
