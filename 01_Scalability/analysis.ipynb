```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
def load_data(file):
    """Returns a numpy array"""
    data = []
    for line in open(file):
        data += [float(line[:-2])]
    return np.asarray(data)
```

## Weak Scaling

```python
weak5 = load_data("parallel/weak_scalability/weak_scal5.dat")
weak6 = load_data("parallel/weak_scalability/weak_scal6.dat")
weak7 = load_data("parallel/weak_scalability/weak_scal7.dat")
```

### Real time execution as a function of the number of cores
The number of core ranges from 1 to 20.

## Strong Scaling

```python
strong5 = load_data("parallel/strong_scalability/strong_scal5.dat")
strong6 = load_data("parallel/strong_scalability/strong_scal6.dat")
strong7 = load_data("parallel/strong_scalability/strong_scal7.dat")
```

```python
# N = 10^5
plt.plot(strong5)
plt.scatter(np.arange(strong5.shape[0]), strong5)

# N = 10^6
plt.plot(strong6)
plt.scatter(np.arange(strong6.shape[0]), strong6)

# N = 10^7
plt.plot(strong7)
plt.scatter(np.arange(strong7.shape[0]), strong7)

plt.show()
```

```python
strong5_speedup = strong5[0] / strong5
strong6_speedup = strong6[0] / strong6
strong7_speedup = strong7[0] / strong7
```

```python
# N = 10^5
plt.plot(strong5_speedup)
plt.scatter(np.arange(strong5_speedup.shape[0]), strong5_speedup)

# N = 10^6
plt.plot(strong6_speedup)
plt.scatter(np.arange(strong6_speedup.shape[0]), strong6_speedup)

# N = 10^7
plt.plot(strong7_speedup)
plt.scatter(np.arange(strong7_speedup.shape[0]), strong7_speedup)
```
